```python
# import libraries
import pandas as pd
import numpy as np
import datetime
import warnings
warnings.filterwarnings("ignore")
```

```python
# 讀取股價變化表，並篩選出目標標的該年度的股價
stock = pd.read_excel('stock_data_2019-2023.xlsx', sheet_name = '上市2021')
corp_stock = stock[stock['證券代碼'] == '3481 群創']
corp_stock['年月日'] = pd.to_datetime(corp_stock['年月日']).dt.date
corp_stock = corp_stock.sort_values(by = '年月日').reset_index(drop = True)
```

```python
corp_stock = corp_stock[['證券代碼', '年月日', '收盤價(元)']]
corp_stock
```

```python
# 計算個股之單日漲跌幅，以及根據規則給予其漲、跌、停的標籤
corp_stock['單日漲跌幅(%)'] = corp_stock['收盤價(元)'].rolling(window=2).apply(lambda x: (x.iloc[1]-x.iloc[0])/x.iloc[0])
corp_stock['label'] = corp_stock['單日漲跌幅(%)'].apply(lambda x: (x>0.005 and '漲') or (x<-0.005 and '跌') or '停')
```
