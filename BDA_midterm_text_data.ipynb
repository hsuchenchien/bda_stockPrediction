```python
# import libraries
import pandas as pd
import numpy as np
import datetime
import warnings
warnings.filterwarnings("ignore")
```

```python
# import 要用來分類的文章檔案，並將其合併
news1921 = pd.read_csv('bda2023_mid_bbs_2019-2021.csv') # 
forum21 = pd.read_csv('bda2023_mid_forum_2021.csv')
news = pd.concat([news1921, forum21]).reset_index().drop(["index"], axis=1)
```

```python
# 透過關鍵字篩選文章檔案
keyWords = ['群創','面板']
corp_news = news[news['title'].str.contains('|'.join(keyWords) +'.*') | news['content'].str.contains('|'.join(keyWords) +'.*')].reset_index(drop = True)
corp_news
```

```python
# import 先前完成之股價漲跌檔案
corp_stock = pd.read_csv('il_stock_2021')
corp_stock['年月日'] = pd.to_datetime(corp_stock['年月日']).dt.date
corp_stock
```

```python
# 將D-1當日的內容標示為該標籤，意即若今天為漲，則將昨日的文章內容標示為漲
corp_news['post_time'] = pd.to_datetime(corp_news['post_time']).dt.date
corp_stock['年月日-1'] = corp_stock['年月日'] - datetime.timedelta(days=1)
corp = pd.merge(corp_news, corp_stock, left_on='post_time', right_on='年月日-1')[['p_type', 'post_time', 'title', 'content', 'label']]
corp
```

```python
# 將結果輸出
corp.to_csv('il_bbsforum2021_ver2', encoding='utf-8')
```
