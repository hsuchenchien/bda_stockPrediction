```python
import pandas as pd
import numpy as np
import datetime
import warnings
warnings.filterwarnings("ignore")
```

```python
# import data
df = pd.read_csv('3481_news_2021')
```

```python
id_content = df['content'].values.tolist()
id_label = df['label'].values.tolist()
id_content_fell = []
id_content_rise = []
id_content_same = []
```

```python
len(id_content)
```

```python
for i in range(len(id_content)):  # 將新聞依當日漲跌分類
    if id_label[i] == '漲':
        id_content_fell.append(id_content[i])
    elif id_label[i] == '跌':
        id_content_rise.append(id_content[i])
    else:  # 在漲跌幅過低的日子，即記錄在id_content_same中
        id_content_same.append(id_content[i])
```

```python
print(len(id_content_fell))
print(len(id_content_rise))
print(len(id_content_same))
```

```python
import monpa
from monpa import utils
import re
```

```python
# 這個function用來將字串以正則化處理去除中文字元以外的字元
def clearSentence(sentence):
    return re.sub(r'[^\u4e00-\u9fa5]+', '', sentence)

# 我們從stopwords_zh.txt這個檔案中匯入繁體中文的停用詞
with open('stopwords_zh.txt', 'r') as file:
    stopwords = file.read().splitlines() 
file.close()
```

```python
# 透過monpa對文章進行斷詞處理，並將個別斷詞結果寫在tokenStr這個字串變數中，最後再將訓練集中所有字串存在tokenStr_list中
train_tokenStr_list = []
for i in range(len(id_content_fell)):
    try:
        if id_content_fell != '':
            sentence_list = utils.short_sentence(id_content_fell[i])
            tokenStr = str()
            for sentence in sentence_list:
                sentence = clearSentence(sentence)
                tokens = monpa.cut(sentence)
                tokenStr += ' '.join(tokens)
            train_tokenStr_list.append(tokenStr)
        if i % 100 == 0:
            print(i)
    except:
        continue
```

```python
for i in range(len(id_content_rise)):
    try:
        if id_content_rise != '':
            sentence_list = utils.short_sentence(id_content_rise[i])
            tokenStr = str()
            for sentence in sentence_list:
                sentence = clearSentence(sentence)
                tokens = monpa.cut(sentence)
                tokenStr += ' '.join(tokens)
            train_tokenStr_list.append(tokenStr)
        if i % 100 == 0:
            print(i)
    except:
        continue
```

```python
for i in range(len(id_content_same)):
    try:
        if id_content_same != '':
            sentence_list = utils.short_sentence(id_content_same[i])
            tokenStr = str()
            for sentence in sentence_list:
                sentence = clearSentence(sentence)
                tokens = monpa.cut(sentence)
                tokenStr += ' '.join(tokens)
            train_tokenStr_list.append(tokenStr)
        if i % 100 == 0:
            print(i)
    except:
        continue
```

```python
# 我們從stopwords_zh.txt這個檔案中匯入繁體中文的停用詞
with open('stopwords_zh.txt', 'r') as file:
    stopwords = file.read().splitlines() 
file.close()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
train_tokenStr_list_train = train_tokenStr_list[:1000]  # 加入跌的訓練資料
train_tokenStr_list_train += train_tokenStr_list[1743:2743]  # 加入跌的訓練資料
train_tokenStr_list_train += train_tokenStr_list[3494:3794]  # 加入漲跌過小的訓練資料
```

```python
train_tokenStr_list_test = train_tokenStr_list[1000:1743]  # 加入跌的測試資料
train_tokenStr_list_test += train_tokenStr_list[2743:3494]  # 加入跌的測試資料
train_tokenStr_list_test += train_tokenStr_list[3794:3881]  # 加入漲跌過小的測試資料
```

```python
vectorizer = TfidfVectorizer(stop_words=stopwords)  # 使用TFIDF判別token重要程度
X_train = vectorizer.fit_transform(train_tokenStr_list_train)
X_train = pd.DataFrame(X_train.toarray(),columns=vectorizer.get_feature_names())
X_train
```

```python
Y_train = []
for i in range(1000):
    Y_train.append('跌')
for i in range(1000):
    Y_train.append('漲')
for i in range(300):
    Y_train.append('停')
```

```python
chi2_selector = SelectKBest(chi2, k = 1000)  # 只取分數最高的1000個 token，去除較不重要的 token
chi2_selector.fit(X_train, Y_train)
kbest_vocabs = X_train.columns[chi2_selector.get_support()]  # 將重要的 token 記錄在 kbest_vocabs 中
X_train = X_train[kbest_vocabs]
X_train
```

```python
from sklearn.svm import SVC
SVM_model = SVC(kernel='linear', C=1.0)
```

```python
SVM_model.fit(X_train,Y_train)  # 模型訓練
```

```python
# 將測試資料依據訓練資料中重要的 token (kbest_vocabs)做 TFIDF 
vectorizer = TfidfVectorizer(stop_words=stopwords)
X_test = vectorizer.fit_transform(train_tokenStr_list_test)
X_test = pd.DataFrame(X_test.toarray(),columns=vectorizer.get_feature_names())
X_test = X_test.reindex(kbest_vocabs, axis=1, fill_value=0)
X_test
```

```python
# 先將訓練資料回丟進模型中，檢測結果
predicted_result = SVM_model.predict(X_train)
expected_result = Y_train
```

```python
from sklearn.metrics import precision_score
precision = precision_score(predicted_result, expected_result, average='micro')
from sklearn import metrics
print(metrics.classification_report(expected_result, predicted_result))
```
